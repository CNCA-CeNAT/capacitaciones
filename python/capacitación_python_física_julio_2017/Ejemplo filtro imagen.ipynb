```python
%matplotlib inline
import numpy as np
from scipy import signal, misc
import matplotlib.pyplot as plt

imagen = misc.face(gray=True)
w = np.zeros((50, 50))
w[0][0] = 1.0
w[49][25] = 1.0
imagen_nueva = signal.fftconvolve(imagen, w)
plt.figure()
plt.imshow(imagen)
plt.gray()
plt.title('Imagen Original')
plt.show()

plt.figure()
plt.imshow(imagen_nueva)
plt.gray()
plt.title("Imagen filtrada")
plt.show()
```
