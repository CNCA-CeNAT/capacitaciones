```r
#1. Leer dataset DATA.csv
mock <- read.csv("Presentaciones_R/dataset/DATA.csv", stringsAsFactors=F)
head(mock)
```

```r
#2. Eliminar registros de China
mockCH <- subset(mock, mock$country != "China")
head(mockCH)
```

```r
#3 ELIMINAR COLUMNAS DE NOMBRE Y APELLIDO------------------------------------------
mockCH <- mockCH[, -c(2,3)]
head(mockCH)
```

```r
#4 Obtener datos estadísticos------------------------------------------------------

#head(mock,18)
mockRe <- mockCH

# Obtener el minimo, maximo, promedio y desviación estándar de la columna age
ageMIN <- min(mockRe$age, na.rm=TRUE)
ageMAX <- max(mockRe$age, na.rm=TRUE)
ageAVG <- mean(mockRe$age, na.rm=TRUE)
ageSD <- sd(mockRe$age, na.rm=TRUE)

#summary(mockCH$age)

# Datos estadísticos----------------------------------------------------------------
# Obtener el minimo, maximo, promedio y desviación estándar de la columna weight
weightMIN <- min(mockRe$weight, na.rm=TRUE)
weightMAX <- max(mockRe$weight, na.rm=TRUE)
weightAVG <- mean(mockRe$weight, na.rm=TRUE)
weightSD <- sd(mockRe$weight, na.rm=TRUE)

#summary(mockCH$weight)

# Obtener el minimo, maximo, promedio y desviación estándar de la columna height
heightMIN <- min(mockRe$height, na.rm=TRUE)
heightMAX <- max(mockRe$height, na.rm=TRUE)
heightAVG <- mean(mockRe$height, na.rm=TRUE)
heightSD <- sd(mockRe$height, na.rm=TRUE)

#summary(mockCH$height)

#Impresión de los datos en una matriz
cat("Datos antes del relleno:")
stat1 <- matrix(c(ageMIN,ageMAX,ageAVG,ageSD,weightMIN,weightMAX,weightAVG,weightSD,
         heightMIN,heightMAX,heightAVG,heightSD),4,3)
colnames(stat1) <- c("Age","Weight","Height")
rownames(stat1) <- c("Min","Max","Mean","SD")
stat1
#--------------------------------------------------------------------------------

# A) RELLENAR VALORES FALTANTES--------------------------------------------------
mockRe$weight[is.na(mockRe$weight)] <- weightAVG
mockRe$height[is.na(mockRe$height)] <- heightAVG
mockRe$age[is.na(mockRe$age)] <- ageAVG

#head(mockRe,18)
#--------------------------------------------------------------------------------


# B) Datos estadísticos----------------------------------------------------------------
# Obtener el minimo, maximo, promedio y desviación estándar de la columna age
ageMIN <- min(mockRe$age, na.rm=TRUE)
ageMAX <- max(mockRe$age, na.rm=TRUE)
ageAVG <- mean(mockRe$age, na.rm=TRUE)
ageSD <- sd(mockRe$age, na.rm=TRUE)

#summary(mockCH$age)

# Obtener el minimo, maximo, promedio y desviación estándar de la columna weight
weightMIN <- min(mockRe$weight, na.rm=TRUE)
weightMAX <- max(mockRe$weight, na.rm=TRUE)
weightAVG <- mean(mockRe$weight, na.rm=TRUE)
weightSD <- sd(mockRe$weight, na.rm=TRUE)

#summary(mockCH$weight)

# Obtener el minimo, maximo, promedio y desviación estándar de la columna height
heightMIN <- min(mockRe$height, na.rm=TRUE)
heightMAX <- max(mockRe$height, na.rm=TRUE)
heightAVG <- mean(mockRe$height, na.rm=TRUE)
heightSD <- sd(mockRe$height, na.rm=TRUE)

#summary(mockCH$height)

#Impresión de los datos en una matriz
cat("Datos después del relleno:")
stat2 <- matrix(c(ageMIN,ageMAX,ageAVG,ageSD,weightMIN,weightMAX,weightAVG,weightSD,
         heightMIN,heightMAX,heightAVG,heightSD),4,3)
colnames(stat2) <- c("Age","Weight","Height")
rownames(stat2) <- c("Min","Max","Mean","SD")
stat2

#--------------------------------------------------------------------------------
```

```r
#PASO 0: Leer y comprender la estructura de los datos
#-------------------------------------------------------------

#Lectura del set de datos
dat <- read.csv("Presentaciones_R/dataset/EconomistData.csv")

#exploración de la estructura del set de datos
str(dat)
head(dat)
```

```r
#PASO 1: Realizar mapeos estéticos
#-----------------------------------------------------------------------
library(ggplot2)

#Mapear CPI al eje "x", HDI al eje "y" y colorear las regiones
pc1 <- ggplot(dat, aes(x = CPI, y = HDI, color = Region))
pc1 + geom_point()
```

```r
##PASO 2: Agregar línea de tendencia
#-----------------------------------------------------------------------

pc2 <- pc1 +
     geom_smooth(mapping = NULL,
                 method = "lm",        #método para ajustar el modelo 
                 formula = y ~ log(x), #log(x), #fórmula a usar poly(x, 2)
                 se = FALSE,           #intervalos de confianza
                 color = "red")

pc2 + geom_point() #Para ver los puntos sobre la linea de tendencia
```

```r
##PASO 3: Estética de los puntos
#-----------------------------------------------------------------------

pc3 <- pc2 +   #cambios en forma y tamaño de los puntos
     geom_point(size = 4.5, shape = 1) + 
     geom_point(size = 4, shape = 1) +
     geom_point(size = 3.5, shape = 1)
pc3


  #Para ver distintos símbolos y formas--------------------------------------
  #df2 <- data.frame(x = 1:5 , y = 1:25, z = 1:25)
  #s <- ggplot(df2, aes(x = x, y = y))
  #s + geom_point(aes(shape = z), size = 4) + scale_shape_identity()  #Observar 25 simbolos de forma
  #s + geom_point(aes(shape = z), size = 4, colour = "Red") + 
  #  scale_shape_identity() #Simbolos en rojo
  #s + geom_point(aes(shape = z), size = 4, colour = "Red", fill = "Black") +
  #  scale_shape_identity() #Simbolos en rojo y con fondo negro
  #---------------------------------------------------------------------------
```

```r
##PASO 4: Etiquetar los puntos
#-----------------------------------------------------------------------

#Lista de paises a etiquetar
pointsToLabel <- c("Russia", "Venezuela", "Iraq", "Myanmar", "Sudan",
                     "Afghanistan", "Congo", "Greece", "Argentina", "Brazil",
                     "India", "Italy", "China", "South Africa", "Spane",
                     "Botswana", "Cape Verde", "Bhutan", "Rwanda", "France",
                     "United States", "Germany", "Britain", "Barbados", "Norway", "Japan",
                     "New Zealand", "Singapore")

#Agregar etiquetas a paises en la lista
pc4 <- pc3 +
    geom_text(aes(label = Country),
              color = "gray20",
              data = subset(dat, Country %in% pointsToLabel))
pc4
```

```r
#Acomodar mejor las etiquetas
#install.packages("ggrepel")
library("ggrepel")
pc4 <-  pc3 +
    geom_text_repel(aes(label = Country),
              color = "gray20",
              data = subset(dat, Country %in% pointsToLabel),
              force = 10)

pc4
```

```r
##PASO 5: Cambiar las etiquetas de las regiones y su orden
#-----------------------------------------------------------------------

datNew <- dat
datNew$Region <- factor(dat$Region,
                       levels = c("EU W. Europe",
                                  "Americas",
                                  "Asia Pacific",
                                  "East EU Cemt Asia",
                                  "MENA",
                                  "SSA"),
                       labels = c("OECD",
                                  "Americas",
                                  "Asia &\nOceania",
                                  "Central &\nEastern Europe",
                                  "Middle East &\nnorth Africa",
                                  "Sub-Saharan\nAfrica"))


pc5 <- pc4
pc5$data <- datNew
pc5

#str(datNew$Region) #variable tipo factor
```

```r
##PASO 6: Nombre y formato del titulo y los ejes
#-----------------------------------------------------------------------


library(grid)
  pc6 <- pc5 +
    scale_x_continuous(name = expression(italic("Corruption Perceptions Index, 2011 (10=least corrupt)")),
                       limits = c(.9, 10.5),
                       breaks = 1:10) +
    scale_y_continuous(name = expression(italic("Human Development Index, 2011 (1=Best)")),
                       limits = c(0.2, 1.0),
                       breaks = seq(0.2, 1.0, by = 0.1)) +
    scale_color_manual(name = "",
                       values = c("#24576D",
                                  "#099DD7",
                                  "#28AADC",
                                  "#248E84",
                                  "#F2583F",
                                  "#96503F")) +
    ggtitle("Corruption and Human development")

pc6
```

```r
##PASO 7: Configuración del tema usando función theme()
#-----------------------------------------------------------------------

pc7 <- pc6 +
    theme_minimal() + # Comenzamos con un tema básico y agregamos gradualmente
    theme(text = element_text(color = "gray20"),
          legend.position = "top", #Posicion de la leyenda arriba
          legend.direction = "horizontal",#Dirección horizontal de la leyenda 
          legend.justification = c(0.1,0),
          legend.text = element_text(size = 11, color = "gray10"),
          axis.text = element_text(face = "italic"),
          axis.title.x = element_text(vjust = -1), # Alejar el titulo del eje x
          axis.title.y = element_text(vjust = 2), # Alejar el titulo del eje y
          axis.ticks.y = element_blank(), # se remueven elementos con la función element_blank()
          axis.line = element_line(color = "gray40", size = 0.5),
          axis.line.y = element_blank(),
          panel.grid.major = element_line(color = "gray50", size = 0.5),
          panel.grid.major.x = element_blank()
    ) + guides(colour = guide_legend(nrow = 1)) #Obliga a la leyenda a quedar en una única línea

mR2 <- summary(lm(HDI ~ log(CPI), data = dat))$r.squared #Calcular R2

  png(file = "economist.png", width = 800, height = 600) #abrir png
  
  pc7

  #Agregar ultimos detalles
  grid.text("Sources: Transparency International; UN Human Development Report",
           x = .02, y = .03,
           just = "left",
           draw = TRUE)
  grid.segments(x0 = 0.81, x1 = 0.825,
                y0 = 0.90, y1 = 0.90,
                gp = gpar(col = "red"),
                draw = TRUE)
  grid.text(paste0("R² = ",
                   as.integer(mR2*100),
                   "%"),
            x = 0.835, y = 0.90,
            gp = gpar(col = "gray20"),
            draw = TRUE,
            just = "left")

  dev.off() #cerrar png
```

```r
#PASO 1 CREAR GRÁFICO DE DISPERSIÓN
#---------------------------------------------------------------------
library(ggplot2)
library(plotly)

mpgPlot <- plot_ly(data = mpg, x = ~class, y = ~cty, type = 'scatter',
  mode = 'markers', color = ~manufacturer, colors = "Set1",
  marker = list(size = 10, line = list(width = 2))) %>%
  layout(title = 'Ahorro de combustible según el vehículo',
         yaxis = list(title='Millas por galón en ciudad'),
         xaxis = list(title='Tipo de vehículo'))

#Exportar gáfico a html
htmlwidgets::saveWidget(mpgPlot, "mpgPlot.html")
```

```r
#PASO 2 AGREGAR MENU DESPLEGABLE CON INFORMACIÓN DEL AÑO
#---------------------------------------------------------------------

#Separar set de datos
mpg1999 <- mpg[mpg$year == 1999,]
mpg2008 <- mpg[mpg$year == 2008,]

mpgPlot2 <- plot_ly() %>%
#Agregamos la traza correspondientes a 1999, con visibilidad verdadera
     add_trace(data = mpg1999, x = ~class, y = ~cty, type = 'scatter', 
               mode = 'markers', color = ~manufacturer, colors = "Set1",
  marker = list(size = 10, line = list(width = 2)),visible=TRUE) %>%

#Agregamos la traza correspondiente a 2008, con visibilidad falsa. 
    add_trace(data = mpg2008, x = ~class, y = ~cty, type = 'scatter', 
               mode = 'markers', color = ~manufacturer, colors = "Set1",
  marker = list(size = 10, line = list(width = 2)),visible=FALSE) %>%

#Para agregar el menú utilizamos el parámetro updatemenus.
     layout(title="Ahorro de combustible según el vehículo",
            xaxis = list(title="Tipo de vehículo"),
            yaxis = list(range = c(6, 36),title="Millas por galón en ciudad", showgrid=FALSE, zeroline=FALSE),
            updatemenus = list(
                list(
                    y = 0.7,
                    buttons = list(
                        list(method = "restyle",
                             args = list("visible", list(TRUE, FALSE)),
                             label = "1999"),
                        
                        list(method = "restyle",
                             args = list("visible", list(FALSE, TRUE)),
                             label = "2008")))))

#Exportar gáfico a html
htmlwidgets::saveWidget(mpgPlot2, "mpgPlot2.html")
```
